```python
#Importing Notebooks
import pandas as pd
import matplotlib.pyplot as plt
#Loading Dataset
D = pd.read_csv('HW3Data.csv',header= None)
D
```

```python
#Selecting rows and columns 
labels = D.loc[:,0].values
features = D.loc[:,1:].values
#Importing train test model to split data into random train test subsets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = \
    train_test_split(features, labels,
                    test_size=0.20,
                    stratify=labels,
                    random_state=100)
```

```python
#Importing support vector for regression and feature ranking with 
#recursive elimination and cross validated selection of best number of features
from sklearn.svm import SVC
from sklearn.feature_selection import RFECV
svc = SVC(kernel = "linear")
rfecv = RFECV(estimator=svc, step=1, cv=5,
             scoring='accuracy')
#Accuracy scoring is proportional to number of correct 
#classifications
rfecv.fit(X_train, y_train)
#Display optimal number of features, figure, and labels
#Plot number of features vs. cross validation scores
print("OPTIMAL NUMBER OF FEATURES: %d"% rfecv.n_features_)
plt.figure()
plt.xlabel("Number of features selected")
plt.ylabel("Cross Validation Score (Number of Correct Classifications)")
plt.plot(range(1, len(rfecv.grid_scores_)+1), rfecv.grid_scores_)
plt.show()
```

```python
#Fitting the svm model to the training data
svc.fit(X_train[:,rfecv.support_], y_train)
#Computing the mean accuracy on the test data and labels
svc.score(X_test[:,rfecv.support_], y_test)
```
